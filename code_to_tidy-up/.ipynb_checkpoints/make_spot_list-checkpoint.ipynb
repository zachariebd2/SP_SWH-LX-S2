```python
def getDateFromStr(N):
    sepList = ["","-","_"]
    YYYYMMDD = ''
    for s in sepList :
        found = re.search('\d{4}'+ s +'\d{2}'+ s +'\d{2}', N)
        if found != None :
           YYYYMMDD = datetime.strptime(found.group(0), '%Y'+ s +'%m'+ s +'%d').date()
           break
    return YYYYMMDD


def getCoords(G):
    
    
    GT = G.GetGeoTransform()
    minx = GT[0]
    maxy = GT[3]
    maxx = minx + GT[1] * G.RasterXSize
    miny = maxy + GT[5] * G.RasterYSize
    
    return minx, maxy, maxx, miny
```
