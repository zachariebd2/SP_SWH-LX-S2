```python
def getDateFromStr(N):
    sepList = ["","-","_","/"]
    date = ''
    for s in sepList :
        found = re.search('\d{4}'+ s +'\d{2}'+ s +'\d{2}', N)
        if found != None :
           date = datetime.strptime(found.group(0), '%Y'+ s +'%m'+ s +'%d').date()
           break
    return date

def getTimeFromStr(N,):
    sepList = ["","-","_"]
    HHMMSS = ''
    for s in sepList :
        found = re.search('-'+'\d{2}'+ s +'\d{2}'+ s +'\d{2}'+'-', N)
        if found != None :
            HHMMSS = datetime.strptime(found.group(0), '%H'+ s +'%M'+ s +'%S'+'_').time()
            break
    return HHMMSS

def reproject(inEPSG,outEPSG,x1,y1):
    
    inProj = Proj(init='EPSG:' + inEPSG)
    outProj = Proj(init='EPSG:'+ outEPSG)
    x2,y2 = transform(inProj,outProj,x1,y1)
    
    return x2, y2

def getCoords(G):
    
    
    GT = G.GetGeoTransform()
    minx = GT[0]
    maxy = GT[3]
    maxx = minx + GT[1] * G.RasterXSize
    miny = maxy + GT[5] * G.RasterYSize
    
    return minx, maxy, maxx, miny

def mkdir_p(dos):
    try:
        os.makedirs(dos)
    except OSError as exc:  # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(dos):
            pass
        else:
            raise
            
def getTileFromStr(N):

    tile = ''
    found = re.search('\d{2}' +'[A-Z]{3}', N)
    if found != None : tile = found.group(0)
    return tile

```

```python
# make list of corresponding dlr and swh images for each tile
dict_landsat_swh = dict()
for tile in LANDSAT_tiles['ALP']:
    dict_landsat_swh[tile] = {}
    minx_S2 = S2_tiles["ALP"][tile]['MINX']
    miny_S2 = S2_tiles["ALP"][tile]['MINY']
    maxx_S2 = S2_tiles["ALP"][tile]['MAXX']
    maxy_S2 = S2_tiles["ALP"][tile]['MAXY']
    epsg_S2 = S2_tiles["ALP"][tile]['EPSG']
    
    minx_S2_4326 , miny_S2_4326 = reproject(epsg_S2,"4326",minx_S2,miny_S2)
    maxx_S2_4326 , maxy_S2_4326 = reproject(epsg_S2,"4326",maxx_S2,maxy_S2)
    wkt_S2 = "POLYGON((%f %f,%f %f,%f %f,%f %f,%f %f))" %  (minx_S2_4326, maxy_S2_4326,
                                                            maxx_S2_4326, maxy_S2_4326,
                                                            maxx_S2_4326, miny_S2_4326,
                                                            minx_S2_4326, miny_S2_4326,
                                                            minx_S2_4326, maxy_S2_4326)
    geom_S2= ogr.CreateGeometryFromWkt(wkt_S2)
    
    for wrs in LANDSAT_tiles['ALP'][tile]:
        dict_landsat_swh[tile][wrs] = {}
        #make landsat dlr list
        landsat_list = glob.glob(os.path.join(LANDSAT_path,'ALP',wrs,'**','*.tif'),recursive=True)
        for landsat in landsat_list:
            text = re.search('[A-Z]{2}\d{2}_[A-Z0-9]{4}_\d{6}_\d{8}_', os.path.basename(landsat)).group(0)
            text = text.split("_")
            platform_landsat=text[0]
            date = text[3]
            year = date[0:4]
            month = date[4:6]
            day = date[6:8]
            
            
            #look for swh of same day for each spot
            for platform_spot in ["SPOT1","SPOT2","SPOT3","SPOT4","SPOT5"]:
                #print(year,month,day,platform_spot)
                swh_list = glob.glob(os.path.join(SWH_path,year,month,day,'**',f'{platform_spot}*.zip'),recursive=True)
                chosen_swh = ""
                max_area=0
                for swh in swh_list:
                    zip_scene = zipfile.ZipFile(swh)
                    MTD_file = ""
                    for ziped_file in zip_scene.namelist():
                        if "MTD_ALL.xml" in os.path.basename(ziped_file) and ".tif.aux.xml" not in ziped_file :
                            MTD_file = ziped_file
                    swh_str = zip_scene.read(MTD_file)
                    tree= ET.fromstring(swh_str)
                    t = tree.findall(".//Product_Characteristics/ACQUISITION_DATE")[0]
                    acqui_date_time = t.text.split("T")
                    t = tree.findall(".//Geoposition_Informations/Geopositioning/Global_Geopositioning/Point[@name='upperRight']/LAT")[0]
                    ur_lat = float(t.text)
                    t = tree.findall(".//Geoposition_Informations/Geopositioning/Global_Geopositioning/Point[@name='upperRight']/LON")[0]
                    ur_lon = float(t.text)
                    t = tree.findall(".//Geoposition_Informations/Geopositioning/Global_Geopositioning/Point[@name='lowerLeft']/LAT")[0]
                    ll_lat = float(t.text)
                    t = tree.findall(".//Geoposition_Informations/Geopositioning/Global_Geopositioning/Point[@name='lowerLeft']/LON")[0]
                    ll_lon = float(t.text)
                    wkt_swh = "POLYGON((%f %f,%f %f,%f %f,%f %f,%f %f))" %  (ll_lon, ur_lat,ur_lon, ur_lat,ur_lon, ll_lat,ll_lon, ll_lat,ll_lon, ur_lat)
                    geom_swh= ogr.CreateGeometryFromWkt(wkt_swh)


                    #we keep the swh product with the most overlap with S2 tile
                    intersection = geom_S2.Intersection(geom_swh)
                    
                    if(intersection is not None and intersection.Area()>max_area): 
                        chosen_swh = swh
                        max_area = intersection.Area()
                
                        
                if chosen_swh != "":
                    print(tile,wrs,date,platform_landsat,platform_spot,acqui_date_time,max_area)
                    if date not in dict_landsat_swh[tile][wrs]:
                        dict_landsat_swh[tile][wrs][date] = {"LANDSAT":{},"SWH":{}}
                    if platform_landsat not in dict_landsat_swh[tile][wrs][date]["LANDSAT"]:
                        dict_landsat_swh[tile][wrs][date]["LANDSAT"][platform_landsat] = landsat
                    if platform_spot not in dict_landsat_swh[tile][wrs][date]["SWH"]:
                        dict_landsat_swh[tile][wrs][date]["SWH"][platform_spot] = chosen_swh   

with open(op.join(PERF_PATH,'overlap_landsat_swh.json'), 'w') as p:
     p.write(json.dumps(dict_landsat_swh))
```

```python
# change Landsat QA values
# nodata: 0 -> 255
# ground: 10 -> 0
# snow : 11 -> 100
# cloud: 12 et 15 (shadows)-> 205
driver = gdal.GetDriverByName('GTiff')
c = 0
for landsat_qa in glob.glob(os.path.join(LANDSAT_path,'**','*.tif'),recursive=True):
    
    text = re.search('[A-Z]{2}\d{2}_[A-Z0-9]{4}_\d{6}_\d{8}_', os.path.basename(landsat_qa)).group(0)
    text = text.split("_")
    Sat=text[0]
    wrs = text[2]
    date = text[3]

    
    
    QA_dir = f"FSC_{date}T000000_{Sat}_{wrs}"
    QA_name = QA_dir+"_FSCTOC.tif"
    QA = gdal.Open(landsat_qa)
    QA_array = QA.ReadAsArray()
    QA_array[QA_array >= 12 ] = 205
    QA_array[QA_array == 0] = 255
    QA_array[QA_array == 10] = 0
    QA_array[QA_array == 11] = 100
    
    
    mkdir_p(os.path.join(LANDSAT_LIS_path,QA_name))
    outDs = driver.Create(os.path.join(LANDSAT_LIS_path,QA_dir,QA_name),  QA.RasterXSize, QA.RasterYSize , 1, gdal.GDT_Byte)
    
    outBand = outDs.GetRasterBand(1)
    outBand.WriteArray(QA_array)
    #set pixel coordinates from the top-left pixel (no need for the odd_margin values)
    outDs.SetGeoTransform(QA.GetGeoTransform()) 
    outDs.SetProjection(QA.GetProjection())
    #save QA raster
    outDs.FlushCache()
    
    c = c + 1
    if c % 50 == 0:
        print(c)
    
    
    
```

```python
# change SWH UNET values
# nodata: 255 -> 255
# ground: 0 -> 0
# snow : 1 -> 100
# cloud: 2 -> 205
#resample to 30m
driver = gdal.GetDriverByName('GTiff')
c = 0
for swh_qa in glob.glob(os.path.join(SWH_path,'**','*SCA.tif'),recursive=True):
    sat = re.search('SPOT\d{1}', os.path.basename(swh_qa)).group(0)
    tile = getTileFromStr(swh_qa)
    dateandtime = (re.search('\d{8}-\d{6}', os.path.basename(swh_qa)).group(0)).split("-")
    date=dateandtime[0]
    time=dateandtime[1]
    
    
    QA_dir = f"FSC_{date}T{time}_{sat}_{tile}"
    QA_name = QA_dir+"_FSCTOC.tif"
    minx = S2_tiles["ALP"][tile]['MINX']
    miny = S2_tiles["ALP"][tile]['MINY']
    maxx = S2_tiles["ALP"][tile]['MAXX']
    maxy = S2_tiles["ALP"][tile]['MAXY']

    mkdir_p(os.path.join(SWH_LIS_path,tile,QA_dir))
    QA = gdal.Open(swh_qa)
    QA = gdal.Warp("",QA,format= 'MEM',resampleAlg="near",xRes=20,yRes=20,outputBounds=[minx,miny,maxx,maxy])
    QA_array = QA.ReadAsArray()
    QA_array[QA_array == 1] = 100
    QA_array[QA_array == 2] = 205
    
    
    
    outDs = driver.Create(os.path.join(SWH_LIS_path,tile,QA_dir,QA_name),  QA.RasterXSize, QA.RasterYSize , 1, gdal.GDT_Byte)
    
    outBand = outDs.GetRasterBand(1)
    outBand.WriteArray(QA_array)
    #set pixel coordinates from the top-left pixel (no need for the odd_margin values)
    outDs.SetGeoTransform(QA.GetGeoTransform()) 
    outDs.SetProjection(QA.GetProjection())
    #save QA raster
    outDs.FlushCache()
    c = c + 1
    if c % 50 == 0:
        print(c)
    
    
```

```python
#generate synthesis
config = {
    "ram": 4096,
    "nb_threads": "8",
    "log_level": "INFO",
}
for year in range(2014,2016,1):
    date1="01/09/"+str(year)
    date1s = str(year)+"0901"
    date2="31/08/"+str(year+1)
    date2s = str(year+1)+"0831"
    for tile in ["31TGL","31TGK"]:
        unet_list_txt = os.path.join(LISTS_path,f"arthur_unet_{tile}.txt")
        landsat_list_txt = os.path.join(LISTS_path,f"arthur_landsat_{tile}.txt")
        with open(unet_list_txt,'r') as unet_file:
            unet_list = unet_file.read().splitlines()
        with open(landsat_list_txt,'r') as landsat_file:
            landsat_list = landsat_file.read().splitlines()
        unet_list_year = []
        for f in unet_list:
            if f"FSC_{year}" in f or f"FSC_{year+1}" in f: unet_list_year.append(f)
        landsat_list_year = []
        for f in landsat_list:
            if f"FSC_{year}" in f  or f"FSC_{year+1}" in f : landsat_list_year.append(f)
            
            
        #generate nodata starter map
        nodata_array = np.ones((5490,5490))*255
        tmp_path = os.path.join(f"/home/ad/barrouz/zacharie/TIMESERIES_PROJECT/ARTHUR_ALPS/SYNTHESIS/{tile}/tmp",f"FSC_{str(year)}0901T000000_NODATA_{tile}")
        mkdir_p(tmp_path)
        outDs = driver.Create(os.path.join(tmp_path,f"FSC_{str(year)}0901T000000_NODATA_{tile}_FSCTOC.tif"),  5490, 5490 , 1, gdal.GDT_Byte)
        geotransform = ([S2_tiles['ALP'][tile]['MINX'], 20, 0, S2_tiles['ALP'][tile]['MAXY'], 0, -20 ])
        outBand = outDs.GetRasterBand(1)
        outBand.WriteArray(nodata_array)
        outDs.SetGeoTransform(geotransform) 
        srs = osr.SpatialReference()
        srs.ImportFromEPSG(int(S2_tiles['ALP'][tile]['EPSG']))
        outDs.SetProjection(srs.ExportToWkt())
        outDs.FlushCache()        
            
        config["tile_id"]=tile
        config["date_start"]=date1
        config["date_stop"]=date2
        config["output_dates_filename"]= f"{date1s}_{date2s}_{tile}.txt"
        config["date_margin"]=30
        
        job_sh_path = "/home/ad/barrouz/WORKSPACE/TIMESERIES_PROJECT/SWH_LANDSAT/synthesis_job.sh"
        #only with swh
        config["input_products_list"]= [tmp_path]
        config["densification_products_list"]= landsat_list_year
        config["config_file"]= f"{CONFIG_path}/landsat_{date1s}_{date2s}_{tile}.json"
        config["output_dir"]= f"/home/ad/barrouz/zacharie/TIMESERIES_PROJECT/ARTHUR_ALPS/SYNTHESIS/{tile}/LANDSAT/landsat_{date1s}_{date2s}_{tile}"
        json_object = json.dumps(config, indent=4)
        with open(config["config_file"], "w") as outfile:
            outfile.write(json_object)
        job_name=f"landsat_{date1s}_{date2s}_{tile}"
        os.system(f"qsub -o {MAIN_path}/logs/log.{job_name} -e {MAIN_path}/logs/log.{job_name} -v config=\"{config['config_file']}\" {job_sh_path}")    
        
        #with usg and landsat
        config["input_products_list"]= unet_list_year
        config["densification_products_list"]= landsat_list_year
        config["config_file"]= f"{CONFIG_path}/swh_landsat_{date1s}_{date2s}_{tile}.json"
        config["output_dir"]= f"/home/ad/barrouz/zacharie/TIMESERIES_PROJECT/ARTHUR_ALPS/SYNTHESIS/{tile}/SWH-LANDSAT/swh_landsat_{date1s}_{date2s}_{tile}"
        json_object = json.dumps(config, indent=4)
        with open(config["config_file"], "w") as outfile:
            outfile.write(json_object)
        job_name=f"swh_landsat_{date1s}_{date2s}_{tile}"
        os.system(f"qsub -o {MAIN_path}/logs/log.{job_name} -e {MAIN_path}/logs/log.{job_name} -v config=\"{config['config_file']}\" {job_sh_path}")        

```

```python
#make graphs
SYNTH_path="/home/ad/barrouz/zacharie/TIMESERIES_PROJECT/ARTHUR_ALPS/SYNTHESIS"

plot_path = "/home/ad/barrouz/zacharie/TIMESERIES_PROJECT/ARTHUR_ALPS/plots"

TCD_max = 15
DTM_range = np.arange(1000,3500,500)

for tile in ["31TGL","31TGK"]:
    
    TCD = (gdal.Open(os.path.join(TCD_PATH,tile,f'TCD_{tile}.tif'))).ReadAsArray()
    DTM = (gdal.Open(glob.glob(os.path.join(DEM_PATH,'**','*'+tile+'*ALT_R2.TIF'),recursive=True)[0])).ReadAsArray()
    for metric in ["NOBS","SCD","SMOD","SOD"]:
        fig, axs = plt.subplots(len(DTM_range),sharex=True)
        for i, DTM_min in enumerate(DTM_range):
            DTM_max = DTM_min + 500
        
            ystart_l, ystart_sl,v_list_l, v_list_sl  = list(),list(),list(),list()
            print(tile,metric)
            period = range(1985,2014)
            for year in period:

                landsat_path = glob.glob(op.join(SYNTH_path,tile,"LANDSAT","**",f"LIS_S2L8-SNOW-{metric}_{tile}_{str(year)}*.tif"),recursive=True)
                swh_landsat_path = glob.glob(op.join(SYNTH_path,tile,"SWH-LANDSAT","**",f"LIS_S2L8-SNOW-{metric}_{tile}_{str(year)}*.tif"),recursive=True)

                if len(landsat_path) == 1:
                    landsat = (gdal.Open(landsat_path[0])).ReadAsArray()
                    landsat = (landsat[(TCD < TCD_max) & (DTM >= DTM_min) & (DTM < DTM_max)]).flatten()
                    v_list_l.append(landsat)
                    ystart_l.append(year)

                if len(swh_landsat_path) == 1:
                    swh_landsat = (gdal.Open(swh_landsat_path[0])).ReadAsArray()
                    swh_landsat = (swh_landsat[(TCD < TCD_max) & (DTM >= DTM_min) & (DTM < DTM_max)]).flatten()
                    v_list_sl.append(swh_landsat)
                    ystart_sl.append(year)



            # regression des medianes
            # median per year
            v_med_l = [np.median(s) for s in v_list_l ]
            v_med_sl = [np.median(s) for s in v_list_sl]
            # Linear trend
            z_l = np.polyfit(ystart_l, v_med_l, 1)
            z_sl = np.polyfit(ystart_sl, v_med_sl, 1)
            p_l = np.poly1d(z_l)
            p_sl = np.poly1d(z_sl)

            space = 0.2
            box_param = dict( widths=0.2, patch_artist=True,
                     flierprops=dict(marker='.', markeredgecolor='black',
                     fillstyle=None), medianprops=dict(color='black'))
            bx1 = axs[i].boxplot(v_list_l, showfliers=False, positions=np.asarray(ystart_l) - space , boxprops=dict(facecolor='tab:blue'),**box_param)
            #ax2 = ax1.twinx()
            bx2 = axs[i].boxplot(v_list_sl, showfliers=False, positions=np.asarray(ystart_sl) + space ,boxprops=dict(facecolor='tab:orange') ,**box_param)
            l1 = axs[i].plot(np.asarray(ystart_l), p_l(ystart_l))
            l2 = axs[i].plot(np.asarray(ystart_sl), p_sl(ystart_sl))


            axs[i].legend([bx1["boxes"][0], bx2["boxes"][0],l1[0], l2[0]], ['LANDSAT', 'SWH + LANDSAT',f"L trend {p_l[1]:.2f} days/year", f"SL trend {p_sl[1]:.2f} days/year"], loc='upper right')

            #ax1.set_ylim(min_v-1,max_v+1)
            #ax2.set_ylim(min_v-1,max_v+1)
            # Format x ticks
            labelsize = 12
            axs[i].set_xticks(period)
            axs[i].set_xticklabels([f'{label}' for label in period],rotation=45)
            axs[i].set_xlabel("YEARS")
            axs[i].set_ylabel("DAYS")
            axs[i].set_title(f"{metric} (days) {tile} TCD < {TCD_max} DTM [{DTM_min},{DTM_max}[")
            axs[i].tick_params(axis='x', labelsize=labelsize)




        
        plt.xticks(rotation=45)
        fig.set_figwidth(20)
        fig.set_figheight(20)
        plt.savefig(op.join(plot_path,f'{metric}_{tile}_TCD_{TCD_max}.pdf'),format="pdf")
```
